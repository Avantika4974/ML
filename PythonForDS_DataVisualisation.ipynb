# **PIE CHARTS**

```python
labels=['php','python','java','c++']
numbers=[75,70,80,60]
fig=plt.figure(figsize=(10,10))
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.pie(numbers,labels=labels,autopct="%0.2f%%",colors=["hotpink","skyblue","teal","#662549"])
axes1.legend()
```

```python
Explode=[0.5,0,0.2,0]
plt.pie(numbers,labels=labels,explode=Explode)
```
