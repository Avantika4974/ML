# **Difference b/w Numpy array and Series**

You can change the indexes of series but not of the numpy array

if number of rows >500K use pandas as it is more efficient else use numpy array

```python
Marks={'maths':90,'physics':92,'chemistry':97,'hindi':98}
ser=pd.Series(Marks)
ser
```

```python
ser=pd.Series(Marks,index=['maths','physics','chemistry'])
ser
```
